```python
import pandas as pd

df = pd.read_csv('data/human_body_temperature.csv')
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

1) Is the distribution of body temperatures normal?
* Although this is not a requirement for CLT to hold (read CLT carefully), it gives us some peace of mind that the population may also be normally distributed if we assume that this sample is representative of the population.

Based on the histogram, the sample data is mound-shaped and slightly skewed left (the mean is a little less than the median; there are also two outliers to the left versus one to the right). The mean of the temperature is 98.25 and the median is 98.30.  The standard deviation is 0.73 and the interquartile range is 0.9.  Based on these values, one can assume that the temperature sample came from a population that is normally distributed.

I used two hypothesis tests that use a null hypothesis that the data sample was drawn from a normal population.  Both p-values were large enough to not reject the null hypothesis.  There is no evidence in the sample data to suggest that it did not come from a normal population.

```python
_ = plt.hist(df.temperature, bins = int(np.sqrt(len(df.temperature))), normed = True)
_ = sns.kdeplot(df.temperature, color = 'red', linestyle = 'dashed')
_ = plt.xlabel('temperature (degrees F)')
_ = plt.ylabel('frequency')
_ = plt.title('Histogram with Kernel Density Estimation')
plt.show()
_ = sns.boxplot(x = 'temperature', data = df)
_ = plt.title('Box Plot of Temperature')
plt.show()
_ = stats.probplot(df.temperature, plot = plt)
plt.show()

print(df.temperature.describe())

W, p = stats.shapiro(df.temperature)
print('Shapiro-Wilk test for normality p-value: ', p)

k2, p = stats.normaltest(df.temperature)
print('Normal test p-value: ', p)
```

3) Is the true population mean really 98.6 degrees F?
* Would you use a one-sample or two-sample test? Why?   
A one-sample test is used here since we are dealing with one sample of temperature data (the 130 observations from the DataFrame).   


* In this situation, is it appropriate to use the *t* or *z* statistic?   
The *t* statistic will be used in this particular case since we are attempting to perform a hypothesis test regarding a mean.  We don't know the true population variance/standard deviation, so a *z* statistic cannot be used.  The sample standard deviation will be used instead, so the *t*-distribution is more appropriate.   


* Now try using the other test. How is the result be different? Why?   
The p-value for the t-test is 2.4106e-07 and the p-value for the z-test is 4.9022e-08.  The tails of a t-distribution are slightly fatter than those of a Normal distribution.  Since the sample size is fairly large (130), the difference between the two distributions is small.  Since the tails in the t-distribution contain more area than the tails in Normal distribution, the p-value will be larger for a t-test.  In effect, the t-test accounts for the fact that we do not actually know the population variance which adds uncertainty.  I created a graph of the Normal distribution and the t-distribution.  The second graph shows a zoomed in section of the left tail.  The white dashed line represents the t-distribution and is above the blue solid which represents the Normal distribution.



```python
t, p = stats.ttest_1samp(df.temperature, 98.6)
print('One sample t-test p-value: ', p)

from statsmodels.stats.weightstats import ztest

z, p = ztest(np.array(df.temperature), value = 98.6)
print('One sample z-test p-value: ', p)
```

```python
mu = df.temperature.mean()
variance = df.temperature.var()
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x,stats.norm.pdf(x, mu, sigma))
plt.plot(x, stats.t.pdf(x, len(x) - 1, mu, sigma), linestyle = 'dashed', color='white')
plt.axis([95.5, 101, 0.0, 0.6])
plt.show()
plt.plot(x,stats.norm.pdf(x, mu, sigma))
plt.plot(x, stats.t.pdf(x, len(x) - 1, mu, sigma), linestyle = 'dashed', color='white')
plt.axis([96.05, 96.1, 0.0, 0.01])
plt.show()
```
